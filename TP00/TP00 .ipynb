Input:

```python
#include <stdio.h>

int main(int argc, int argv[])
{
    printf("%d\n", 1 >> 1);
    printf("%d\n", 1 << 1);
    printf("%d\n", 1 << 2);
    
    return 0;
}
```

The output is supposed to be: 

    00000000000000000000000000000000 
    
    00000000000000000000000000000010

Left shift operator (<<) takes two numbers, left shift the bits(binary representation of that number) of the first operand, the second operand decides the number of places to shift.
When data is shifted left, trailing digits are filled with zero.

Right shift operator (>>) takes two numbers, right shift the bits (binary representation of that number) of the first operand, the second operand decides the number of places to shift.
When data is shifted left, digits from right are filled with zero.

which can explain the results above.

Shifting left by 1 bit is the same as multiplying the number by 2. There is a direct analogous when we work with base 10. If we take the number 3 in base 10 and we shift it left, we get 30 which is 3 * 10, and the factor 10 appears because we are working with base 10.

The same applies to base 2. Shifting left is the same as multiplying by 2.

When we write a number like this: 

                    a(n)a(n-1)...a(2)a(1)a(0) in base b, 
                    
what we actually mean is the following :

         a(n)a(n-1)...a(2)a(1)a(0)0 
         
         = a(n).b^n-1 + a(n-1).b^n + ... + a(2).b^3 + a(1).b^2 + a(0).b + 0.b^0
         
         =b(a(n).b^n + a(n-1).b^n-1 + ... + a(2).b^2 + a(1).b + a(0).b^0)
         
         =b(a(n)a(n-1)...a(2)a(1)a(0))
         
         
                

With this in mind one can show that the two operations of shifting left and multiplying by b are actually the same.
